**df.explode('job_skills')**

```python
# Import Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (to convert job_posted_date to datetime)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

The .**explode()** method will help us clean up the dataset a bit further

Let's use a simple example with a dataset of 3 job postings.

```python
data = {
    'job_title_short': ['Data Analyst', 'Data Scientist', 'Data Engineer'],
    'job_skills': [['excel', 'sql', 'python'], ['python', 'r'], ['aws', 'python', 'airflow']]
}

df_skills_lists = pd.DataFrame(data)

df_skills_lists
```

```python
# Using the explode method, each job_title_short is repeated for each skill
# Note that because there are 3 skills for Data Analyst, the job_title_short is repeated 3 times
df_skills_lists.explode('job_skills')
```

```python
# Now we could you the value_counts() method to count the number of times each skill appears
df_skills_lists.explode('job_skills').value_counts('job_skills')
```

```python
# ...and of course, we can plot it!
df_skills_lists.explode('job_skills').value_counts('job_skills').plot(kind='barh')
```

## Pandas - explode()

Using our dataset

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x) #Cleanup the job_skills column into a list data type
```

```python
df[['job_title_short','job_skills']].head()
```

```python
help(df.explode)
```

```python
df_exploded = df.explode('job_skills')
```

```python
df_exploded
```

```python
df_exploded['job_skills'].value_counts().head(10)
```

```python
df_exploded['job_skills'].value_counts().head(10).plot(kind='bar')
```

# Categorizing

Not only by skills, but also by job_title_short

```python
df_exploded.groupby(['job_skills', 'job_title_short']).size() #using size() so we can see the counts
```

```python
df_exploded.groupby(['job_title_short', 'job_skills']).size()
```

```python
skills_count = df_exploded.groupby(['job_title_short', 'job_skills']).size()
# skills_count is a Series, so we can convert it to a DataFrame
df_skills_count = skills_count.reset_index(name='skill_count')
df_skills_count = df_skills_count.sort_values(by='skill_count', ascending=False)
df_skills_count
```

```python
job_title = 'Data Analyst'
top_skills = 10

df_skill_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)

df_skill_final
```

```python
# Let's plot this bad boy!

# Because this data set has many variables, we are going to point out the x and y axis
df_skill_final.plot(kind='barh', x='job_skills', y='skill_count')
```

```python
#  Now, we want to invert the order of the y axis. We could go back and change the ascending order and instead of selecting head(top_skills), we could select tail(top_skills) but that would be a pain
# So, we're going to use the matplotlib.axes.Axes.invert_yaxis() method
# ...and to get the Axes, we need to run the method matplotlib.pyplot.gca() (get current Axes)
df_skill_final.plot(kind='barh', x='job_skills', y='skill_count')
plt.gca().invert_yaxis()
plt.title(f'Top {top_skills} Skills for {job_title}s')
plt.xlabel('Job Posting Count')
plt.ylabel('') #removing the y axis label, since it's self-explanatory
plt.legend().remove() #removing the legend, since it's self-explanatory
plt.show()
```
