# Pandas Accessing Data Problems

```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (to convert job_posted_date to datetime)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Select First 5 Rows with .iloc() (2.1.1) - Problem

Select and display the **job_title** and **job_location** columns for the first 5 rows of the DataFrame using the iloc method.

Hint:

Use the **iloc** method to select the first 5 rows and specific columns by their index positions.

```python
df.iloc[0:5,1:3]
# Alternatively df.iloc[:5,[1,2]]
```

## Get Rows and Columns with .loc() (2.1.2) - Problem

Retrieve the columns starting with **job_title** all the way to **job_posted_date** for rows with an index of 5 to 9. Use the **loc** method.

Hint:

Use the slicing with the **.loc()** method to get the rows with index 5 to 9 and the appropriate columns. 


```python
df.loc[5:9, 'job_title':'job_posted_date']
```

## First 15 Rows in US (2.1.3) - Problem

Retrieve the **job_title** and **salary_year_avg** for the first 15 rows where the **job_country** is 'United States'. Use the **iloc** method.

Hint:
Use boolean indexing to filter the rows where **job_country** is 'United States' and then use **iloc** to get the first 15 rows.

```python
df_us = df[df['job_country'] == 'United States']
df_us[['job_title','salary_year_avg']].iloc[:15]
```

## Filter by Date (2.1.4) - Problem

Select and display the **job_title**, **job_location**, and **job_posted_date** columns for rows where the **job_posted_date** is after '2023-05-01' using the **loc** method.

Use the **loc** method to filter rows based on the **job_posted_date** column and select specific columns by their labels.
