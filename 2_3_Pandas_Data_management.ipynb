```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (to convert job_posted_date to datetime)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### df.sample()

**The df.head() and the df.tail() methods show the first 5 and last 5 rows of the dataset, but they're not really representative of our data**

**To get a random sample of our data, we can use the method df.sample()**

```python
df.sample(10)
# by looking at the index column, we can tell that this method returned random rows.
# Every time we run this method, it will provide a different set of rows
```

**If you have a result of a sample that you want to keep, you can run the sample method, with this parameter**
**It will ensure that the output result will always be the same rows**

```python
df.sample(10, random_state=42)
```

### df.copy()

**When we assign a dataframe to another dataframe, they both keep the same id, and that is a problem because as we alter one, the other also gets altered**

**to prevent this, we use the .copy() method in order to create a dataframe with a different id**

```python
# Let's create a new dataframe called df_original for this example
df_original = dataset['train'].to_pandas()
# Let's create another dataframe from the df_original
df_new = df_original
# Now let's check if they share the same id
id(df_original) == id(df_new)
```

**Now, if we use the copy() method, we can create a copy of the dataframe, with another id**

**This is important because if we alter something in one dataframe, we will not alter the other**


```python
df_altered = df_original.copy()
# Now let's check their ids
id(df_altered) == id(df_original)
```
