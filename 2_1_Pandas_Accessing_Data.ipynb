# Pandas Accessing Data

```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (to convert job_posted_date to datetime)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Using iloc[] to look at the dataset rows
- we can use df.iloc[0] to see only the first row
- we can use df.iloc[0:10] to see the first 10 rows
- we can use df.iloc[0:10,0] to see the only the first column of the first 10 rows 

```python
df.iloc[0:10,0]
```

Another way to see data for certain columns, and with the bonus of not having to know what their index number is, we can alternatively use df.loc[]
- we can use df.loc[0:10, 'salary_rate'] to see only the column 'salary_rate' of the first 10 rows
- we can use slicing to get more than one column: df.loc[0:10, 'salary_rate': 'salary_hour_avg']


```python
df.loc[0:10, 'salary_rate': 'salary_hour_avg']
```

```python
# To get all rows, we use the following
df.loc[:, 'salary_rate': 'salary_hour_avg']
```

```python
# It's not very useful data as it's showing a bunch of None and NaN values.
# We can filter this by using the command dropna on the salary_rate column
df.loc[:, 'salary_rate': 'salary_hour_avg'].dropna(subset='salary_rate')
```
