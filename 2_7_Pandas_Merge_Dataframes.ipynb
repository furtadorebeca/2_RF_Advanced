```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (to convert job_posted_date to datetime)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

from time to time we will find out that our data is incomplete, so we will need to merge it with new data.

First, for simplicity, we're going to use a dataset similar to our already familiar df as the *left* DataFrame, called *job_data*.

The *right* DataFrame is the object we're going to merge with it. We're also going to use a simple one called *company_data*.

**Note:** We need to specify which column/field we will use from each DataFrame when merging them. If the fields in both DataFrames have the same name, then we will use the **on** parameter. If they however have different names we would use two parameters: **left_on** and **right_on**

```python
job_data = {
    'job_id': [1, 2, 3, 4, 5],
    'job_title': ['Data Scientist', 'Software Engineer', 'Product Manager', 'Marketing Director', 'HR Manager'],
    'company_name': ['Google', 'Microsoft', 'Apple', 'Nike', 'Starbucks'],
    'job_location': ['Mountain View, CA', 'Redmond, WA', 'Cupertino, CA', 'Beaverton, OR', 'Seattle, WA']
}

company_data = {
    'company_name': ['Google', 'Microsoft', 'Apple', 'Nike', 'Starbucks'],
    'industry': ['Technology', 'Technology', 'Technology', 'Apparel', 'Food & Beverage'],
    'company_size': ['100,000+', '100,000+', '100,000+', '75,000+', '346,000+']
}


df_jobs = pd.DataFrame(job_data)
df_companies = pd.DataFrame(company_data)
```

```python
df_jobs.head()
```

```python
df_companies.head()
```

```python
# Merging the DataFrames
df_jobs.merge(df_companies, on='company_name') 
```

We can see above that the DataFrames were merged into the job_id DataFrame.

Now, let's gmove to a more realist example. We're going to use the DataFrame from a previous exercize where we created a pivot table with the number of job posts per month for each hob title.

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot = df_US_pivot.reset_index()
df_US_pivot['job_posted_month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot = df_US_pivot.sort_values('job_posted_month_no')
df_US_pivot = df_US_pivot.set_index('job_posted_month')
df_US_pivot = df_US_pivot.drop(columns='job_posted_month_no')

df_US_pivot
```

The DataFrame above only shows Data type of jobs. There is a field commonly associated to **Data Science** and that is **Software Development**.

Let's make some fake data for this.

```python
# ***** NOTE: The code here is not important for you to understand at this time. *****

import random

# Set seed for reproducibility
random.seed(42)

# Define job titles
job_titles = ['Front-End Developer', 'Back-End Developer', 'Full-Stack Developer', 'UI/UX Designer']

# Define months
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Define ranges for random data
ranges = {
    'Front-End Developer': (11000, 15000),
    'Back-End Developer': (8000, 10000),
    'Full-Stack Developer': (5000, 7500),
    'UI/UX Designer': (4000, 5000)
}

# Generate random data
data = {}
for job_title in job_titles:
    data[job_title] = [random.randint(ranges[job_title][0], ranges[job_title][1]) for _ in months]

# Create DataFrame
df_US_software_pivot = pd.DataFrame(data, index=months)
df_US_software_pivot.index.name = "job_posted_month"

df_US_software_pivot
```

Luke also made this Data available in a website and we can get it from there with this Pandas method *.read_csv()*

The only caveat is that this DataFrame contains an index column and we want it to rather use the job_posted_month column as the index. For this, we're going to use the parameter *index_col='job_posted_month'*

In the code below, we will assign this DataFrame to a vaiable called *df_US_software_pivot*.

```python
df_US_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
df_US_software_pivot
```

Now let's merge these 2 DataFrames!

```python
df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

And now we can see that we have Data Science jobs numbers as well as Software Development ones.


In the .merge() method, we have a parameter called **how** that by default is set to **inner**

Let's now create a plot of the top 5 jobs in the US. 

For this, we're going to assign our merged DataFram to the variable *df_US_merged*

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
# PRO TIP: Because we have many methods here, we can put them all within parenthesis so we can separate in different lines. 
# It makes it easier to read and understand!

top_5 = (
    df_US_merged
    .sum() # With sum, we're summing the columns so we have the total counts of all these things
    .sort_values(ascending=False)
    .head(5) # the number 5 here is optional, because head() already returnes the first 5 results
    .index # because we're not looking for the numbers in here, but only the index, we added the .index method
    .to_list() #this transform the result into a list
) 

top_5
```

Now let's plot this DataFrame!

```python
df_US_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 22000) # this is to make the y limit higher do that the legend doesn't appear on top of the chart
plt.legend()
plt.show()
```

The chart shows us that there are more *Front-End Developer* and *Back-End Developer* postings, which is expected to see.
