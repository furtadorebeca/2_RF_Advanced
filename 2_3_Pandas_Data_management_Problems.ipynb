```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (to convert job_posted_date to datetime)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### Drop Missing Values (2.3.1) - Problem

Create a copy of the DataFrame and drop rows where the **salary_year_avg** column has missing values. Display the first 5 rows of the cleaned DataFrame.

Hint:
Use the **copy()** method to create a copy of the DataFrame and the **dropna()** method to drop rows with missing values.

```python
df_copy = df.copy()
df_cleaned = df_copy.dropna(subset=['salary_year_avg'])
df_cleaned.head(5)
```

### Random Sample of DataFrame (2.3.2) - Problem

Create a new DataFrame that is a random sample of 10% of the original DataFrame. Display the first 5 rows of the sampled DataFrame.

Hint:

Use the **sample()** method to create a random sample of the DataFrame.

```python
df_sample = df.sample(frac=0.1)
df_sample.head(5)
```

## Fill Missing Salaries (2.3.3) - Problem

Create a copy of the original DataFrame and fill in missing values in the **salary_year_avg** column with the median salary. Display the first 5 rows of the altered DataFrame.

Hint:

Use the **copy()** method to create a copy of the DataFrame and the **fillna()** method to fill in missing values.

```python
df_copy = df.copy()
median_salary = df_copy['salary_year_avg'].median()
df_copy['salary_year_avg'] = df_copy['salary_year_avg'].fillna(median_salary)
```
