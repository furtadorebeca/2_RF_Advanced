# Pandas Explode Exercise - Investigate Trending Skills

**df.explode('job_skills')**

```python
# Import Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (to convert job_posted_date to datetime)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

For this exercise we're going to focus on the **Data Analyst** position.

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA
```

```python
# Extract the month out of the job_posted_date
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
```

```python
df_DA
```

```python
df_DA_explode = df_DA.explode('job_skills')
```

```python
df_DA_explode
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0) # the parameter fill_value replaces the NaN values with 0
```

```python
df_DA_pivot
```

```python
df_DA_pivot.plot(kind='line', figsize=(20, 10), title='Data Analyst Job Skills by Month', xlabel='Month', ylabel='Number of Job Postings')
plt.xticks(rotation=0)
plt.legend(title='Job Skills', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Create a row that sums up all the values

```python
# Let's create a row that sums up all the values in the pivot table
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
```

```python
df_DA_pivot.loc['Total'].sort_values(ascending=False).index
```

```python
# With this we sort the pivot table by the total number of job postings for each skill. starting with the most common skills
df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
# Now, let's drop the Total row
df_DA_pivot = df_DA_pivot.drop('Total')
```

```python
df_DA_pivot
```

```python
# NOw let's get the top 5 skills
df_DA_pivot_top5 = df_DA_pivot.iloc[:, :5]
```

```python
df_DA_pivot_top5
```

```python
# Let's plot this bad boy!
df_DA_pivot_top5.plot(kind='line', figsize=(20, 10), title='Top 5 Skills for Data Analyst by Month', xlabel='', ylabel='Count')
```
